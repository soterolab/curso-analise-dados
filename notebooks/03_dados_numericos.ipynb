###  As msicas mais longas so as mais populares?

```python
dfSpotify.columns
```

```python
dfSpotify[['duration_ms', 'popularity']].sort_values('duration_ms', ascending=False).head(5)
```

```python
dfSpotify['duracao_min'] = dfSpotify['duration_ms'] / 60000
dfSpotify.head()
```

```python
dfSpotify.dtypes
```

```python
dfSpotify.loc[dfSpotify['duracao_min'] <= 1.0, 'faixa_duracao'] = 'menor que 1 min'
```

```python
dfSpotify.loc[(dfSpotify['duracao_min'] > 1.0) & (dfSpotify['duracao_min'] <= 2.0), 'faixa_duracao'] = 'entre 1 e 2 min'
```

```python
dfSpotify.loc[dfSpotify['duracao_min'] <= 1.0, 'faixa_duracao'] = 'menor que 1 min'
dfSpotify.loc[(dfSpotify['duracao_min'] > 1.0) & (dfSpotify['duracao_min'] <= 2.0), 'faixa_duracao'] = 'entre 1 e 2 min'
dfSpotify.loc[(dfSpotify['duracao_min'] > 2.0) & (dfSpotify['duracao_min'] <= 3.0), 'faixa_duracao'] = 'entre 2 e 3 min'
dfSpotify.loc[dfSpotify['duracao_min'] > 3.0, 'faixa_duracao'] = 'maior que 3 min'
dfSpotify.head(5)
```

```python
dfSpotify['faixa_duracao'].value_counts()
```

```python
dfSpotify[['faixa_duracao', 'popularity']].groupby('faixa_duracao').sum('popularity').reset_index().sort_values('popularity', ascending=False).tail(5)
```

```python
dfSpotify[['faixa_duracao', 'popularity']].groupby('faixa_duracao').mean('popularity').reset_index().sort_values('popularity', ascending=False).tail(5)
```
