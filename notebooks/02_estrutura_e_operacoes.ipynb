````python
dfConversao = pd.read_csv(lista_conversao, sep=',')
dfConversao
````

````python
df_merged = pd.merge(df, dfConversao, on='País', how='inner')
df_merged
````

````python
df_merged['Valor convertido'] = df_merged['Último'] * df_merged['Conversão BR']
df_merged
````

````python
df_merged[['País', 'Último', 'Valor convertido']]
````

````python
df_merged.sort_values('Valor convertido', ascending=False)
````

````python
dfSalariosMes = df_merged[df_merged['Unidade'].str.lower().str.contains('mês')]
dfSalariosMes
````

````python
dfSalariosMes.sort_values('Valor convertido', ascending=False)
````

````python
dfSalariosMes['Moeda'] = dfSalariosMes['Unidade'].str.split('/', expand=True)[0]
dfSalariosMes
````

````python
dfSalariosMes[['País', 'Valor convertido', 'Moeda']].groupby('Moeda').count().reset_index().sort_values('Valor convertido', ascending=False).head(3)
````
